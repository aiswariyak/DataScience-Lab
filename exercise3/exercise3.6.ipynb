```python
import matplotlib.pyplot as plt
#3.6.1
data={'java':22.2,'python':17.6,'PHP':8.8,'JavaScript':8,'C#':77,'C++':6.7}
xdata=list(data.keys())
ydata=list(data.values())
plt.bar(xdata,ydata)
plt.xlabel("programming languages")
plt.ylabel("popularity")
plt.show()
#3.6.2
plt.barh(xdata,ydata,color="red")
plt.ylabel("programming languages")
plt.xlabel("popularity")
plt.title("Horizondal barchart")
plt.show()
#3.6.3
plt.bar(xdata,ydata,color=('r','g','b','gray','y','orange'))
plt.ylabel("programming languages")
plt.xlabel("popularity")
plt.show()
```
